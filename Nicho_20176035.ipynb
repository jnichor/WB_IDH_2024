```python
import pandas as pd
import altair as alt
```

```python
base = pd.read_excel("./IDH_PNUD_2017-1.xlsx", sheet_name= "Departamental", header=2)
base
```

```python
base.info()
```

```python
# Eliminamos columnas que en blanco
base = base[["UBIGEO","DEPARTAMENTO", "Población", "Unnamed: 6", "Índice de Desarrollo Humano", "Unnamed: 8","Esperanza de vida al nacer","Unnamed: 10","Con Educación secundaria completa (Poblac. 18 años)", "Unnamed: 12", "Años de educación (Poblac. 25 y más)", "Unnamed: 14","Ingreso familiar per cápita", "Unnamed: 16"]]
base
```

```python
#Eliminamos filas vacías
base = base.drop([0, 1, 2, 3, 30, 31,32])
base
```

```python
# Reseteamos el index
base = base.reset_index(drop=True)
base
```

```python
base.info()
```

```python
# Cambio de nombre de las columnas
def renombrar_columnas():   
    '''
    cambia el nombre de las columnas de forma directa y menciona 'listo' al hacerlo
    Inputs:
        No hay input en la función
    Output:
        Retorna: la palabra listo para asegurar que se ha ejecutado la funcion sin parametros
    ''' 
    nuevos_nombres = {'UBIGEO': 'Ubigeo', 'DEPARTAMENTO': 'Departamento','Unnamed: 6':'Ranking_hab', 'Unnamed: 8': 'Ranking_IDH', 'Unnamed: 10': 'Ranking_Esperanza', 'Unnamed: 12': 'Ranking_Secundaria','Unnamed: 14': 'Ranking_educacion', 'Unnamed: 16': 'Ranking_ingresos'}

    base.rename(columns = nuevos_nombres, inplace = True)
    print("Listo")

renombrar_columnas()

base
```

```python
base.info()
```

```python
#Identificamos la cantidad missings por columna
base.isna().sum()
```

```python
alt.Chart(base).mark_point()
```

```python
# Elimine las observaciones missings values
base= base.dropna()
base
```

```python
# Elegimos un tema predeteminado de altair
alt.themes.enable('opaque')
```

```python
alt.Chart(base).mark_point()
```

```python
# Grafico 1
alt.Chart(base, height=500, width=500).mark_point(color='darkblue').encode(
y = "Departamento",
x = "Índice de Desarrollo Humano",
size = alt.Size("Ingreso familiar per cápita"),
tooltip = ["Departamento", "Ingreso familiar per cápita", "Índice de Desarrollo Humano", "Ranking_IDH"]
).interactive().properties(
    title="Distribución del Índice de Desarrollo Humano por departamento en Perú",

)
```

```python
alt.Chart(base).mark_bar()
```

```python
# Grafico 2
alt.Chart(base, height=400, width=750).mark_bar(color='darkblue').encode(
x = "Departamento",
y = "Población",
tooltip = ["Departamento", "Ingreso familiar per cápita","Población", "Ranking_hab"]
).interactive().properties(
    title="Distribución de habitantes por cada departamento en Perú",
)
```
